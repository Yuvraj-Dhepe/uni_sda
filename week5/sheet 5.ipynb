```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler

# Create the dataset
data = {
    'house_price': [200000, 220000, 250000, 180000, 210000],
    'square_footage': [1500, 1800, 2200, 1200, 1600],
    'num_bedrooms': [3, 4, 5, 2, 3],
    'distance_from_city': [2, 1, 0, 3, 2]
}

df = pd.DataFrame(data)
```

```python

# Split the dataset into features (X) and target variable (y)
X = df[['square_footage', 'num_bedrooms', 'distance_from_city']]
y = df['house_price']
```

```python

# Fit a linear regression model
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python

# Print the coefficients
print("Linear Regression Coefficients:")
print("Intercept:", linear_model.intercept_)
print("Coefficients:", linear_model.coef_)
```

```python

# Fit a ridge regression model with varying regularization parameter (Î»)
alphas = np.arange(0, 1.01, 0.01)
ridge_coeffs = []
```

```python
for alpha in alphas:
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X, y)
    ridge_coeffs.append(ridge_model.coef_)
```

```python
# Plot the coefficients as a function of Î»
ridge_coeffs = np.array(ridge_coeffs)
plt.figure(figsize=(10, 6))
plt.plot(alphas, ridge_coeffs)
plt.title('Ridge Regression Coefficients vs. Regularization Parameter')
plt.xlabel('Regularization Parameter (Î»)')
plt.ylabel('Coefficients')
plt.legend(['Square Footage', 'Number of Bedrooms', 'Distance from City'])
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Creating a DataFrame from the given data
data = {
    'house_price': [200000, 220000, 250000, 180000, 210000],
    'square_footage': [1500, 1800, 2200, 1200, 1600],
    'num_bedrooms': [3, 4, 5, 2, 3],
    'distance_from_city': [2, 1, 0, 3, 2]
}

df = pd.DataFrame(data)

# Separating the predictors (features) and the target variable
X = df[['square_footage', 'num_bedrooms', 'distance_from_city']]
y = df['house_price']

# Fitting the linear regression model
model = LinearRegression()
model.fit(X, y)

# Getting the coefficients of the model
intercept = model.intercept_
coefficients = model.coef_

# Printing the desired output
print(f'Intercept (beta0): {intercept}')
print(f'Coefficient for sq_footage (beta1): {coefficients[0]}')
print(f'Coefficient for num_bedrooms (beta2): {coefficients[1]}')
print(f'Coefficient for distance_from_city (beta3): {coefficients[2]}')
```

```python
import numpy as np

# Given coefficients
beta = np.array([54999.99999999997, 100.00000000000004, -5000.000000000006, 5000.000000000005])

# New data
new_sq_footage = np.array([1500])
new_num_bedrooms = np.array([3])
new_distance_from_city = np.array([2])

# Create the design matrix for new data
X_new = np.column_stack((np.ones_like(new_sq_footage), new_sq_footage, new_num_bedrooms, new_distance_from_city))

# Use the final coefficients to make predictions on new data
predictions_new = X_new @ beta

# Print the predictions for new data
print("Predictions for new data:", predictions_new)
```
