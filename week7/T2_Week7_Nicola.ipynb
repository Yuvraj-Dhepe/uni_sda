```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def lrt(t0, min_x, n):
    return np.exp(n * (t0 - min_x) )if t0 <= min_x else 1

def test_t0(t0, minima, n):
    res = np.zeros(len(minima))

    for i, min_x in enumerate(minima):
        res[i] = lrt(t0, min_x, n)

    return  res
```

```python
n = 100
minima_x = np.linspace(-10,10,n)
theta_0_vals = [-2, 0, 0.5, 1]
```

```python
for t0 in theta_0_vals:
    res=test_t0(t0, minima_x, n)
    plt.plot(minima_x, res, label=f'theta_{0}={t0}')
    plt.xlabel('$\min_{i\in\{1,\dots,n\}}{x_i}$')
    plt.ylabel('$\lambda(x_1,\dots,x_n)$')
    plt.legend(loc='best')
    plt.title(f'LRT statistic for different $\\theta_{0}$ values and n={n}')

plt.show()
```
