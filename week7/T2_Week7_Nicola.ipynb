```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate samples
np.random.seed(7113)
theta = np.random.normal()
samples = [np.random.exponential(scale=1, size=100)]
```

```python
def lrt(t0, x):
    return np.exp(len(x) * (t0 - np.mean(x)))
```

```python
theta_0_vals = [-2, 0, 0.5, 1]
for t0 in theta_0_vals:
    print(f"{t0}: LRT {lrt(t0, samples)}")
```

```python
lrts = [lrt(t0, samples) for t0 in theta_0_vals]
plt.plot(theta_0_vals, lrts)
```
