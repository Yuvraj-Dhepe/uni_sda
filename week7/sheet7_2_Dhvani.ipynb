```python
import numpy as np
import matplotlib.pyplot as plt

def likelihood_ratio_test(x_values, theta0):
    """
    Computes the likelihood ratio test statistic for the given dataset and theta0.

    Args:
        x_values (numpy.ndarray): The dataset of independent random variables.
        theta0 (float): The value of theta0 for the hypothesis test.

    Returns:
        float: The likelihood ratio test statistic.
    """
    # Compute the likelihood of the observed data under the null hypothesis
    L_null = np.prod([np.exp(-(xi - theta0)) for xi in x_values if xi >= 0])

    # Compute the likelihood of the observed data under the alternative hypothesis
    L_alt = np.prod([np.exp(-xi) for xi in x_values])

    # Compute the likelihood ratio test statistic
    lr_statistic = np.log(L_alt) - np.log(L_null)

    return lr_statistic

# Generate some sample data
n = 100
x_values = np.random.exponential(1, size=n)

# Define the values of theta0 for which we want to compute the LRT statistic
theta0_values = np.array([-2, 0, 0.5, 1])

# Compute and plot the LRT statistic for each theta0 value
lr_statistics = []
for theta0 in theta0_values:
    lr_statistics.append(likelihood_ratio_test(x_values, theta0))

# Plot black dots
plt.plot(theta0_values, lr_statistics, marker='o', color='black')


# set limits slightly higher than extent
plt.xlim(theta0_values.min() - 0.1, theta0_values.max() + 0.1)

# set labels
plt.xlabel('Theta0')
plt.ylabel('Likelihood ratio test statistic')
plt.title('LRT statistic for different theta0 values')
plt.show()
```
